# Pyplot Tutorial 
taken from http://matplotlib.org/users/pyplot_tutorial.html

```python
import matplotlib.pyplot as plt
import numpy as np
# since only one list is passed in, it is assumed to be the y-axis
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')

# To plot x versus y, you can do
plt.plot([1, 2, 3, 4], [1, 4,9, 16])

# you can pass a format string which specifies line color and style
# to plot the above with red circles, you can do
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

## Figures and Subplots

```python
fig = plt.figure()

# Add a 2x2 plot, selecting the first of 4 subplots
ax1 = fig.add_subplot(2, 2, 1)

# we can make the next two plots as well
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
# Now we have a matplotlib Figure with 3 empty subplots

# if we use plt.plot(), it draws on the last figure and subplot used
from numpy.random import randn
plt.plot(randn(50).cumsum(), 'k--')
# k-- is a style option for black dashed line

ax1.hist(randn(100), bins=20, color='k', alpha=.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))
```
