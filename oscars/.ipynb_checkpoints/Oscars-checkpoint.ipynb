# Analyzing the Oscars Dataset
Learn things about the oscars and when they were given

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv("database.csv")
df = df.set_index('Year')
df.head()
```

```python
# Let's count how many awards were given at each ceremony
plt.figure(figsize=(15, 10))
awards = df['Winner'].groupby(df['Ceremony']).sum()
awards.plot()
plt.title("Number of Awards Given")
plt.ylabel("# Awards")

# Max and min # awards
yearmax = np.argmax(awards) + 1
plt.scatter(yearmax, awards.iloc[yearmax], 100, color="green")
plt.plot([yearmax], [0, awards[yearmax]], 'b--')
df.index.unique()[yearmax+1]
```
